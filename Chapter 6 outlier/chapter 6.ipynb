# standard deviation mathod

```python
# in this section we calculate the mean and standard deviation and 
# 1 standard deviation from the mean covers 68% of the data
# 2 standard deviation from the mean covers 95% of the data
# 3 standard deviation from the mean covers 99.7% of the data 
```

```python
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import std
seed(1)
data= 5 * randn(10000) + 50

# calculate statistics
data_mean, data_std= mean(data), std(data)

#  define outliers
cut_off=data_std*3
lower, upper= cut_off-data_mean, cut_off+data_mean

# identify outliers
outliers=[x for x in data if x <lower or x> upper]
print("outliers is:{}" .format (len(outliers)))

# remove outliers
outliers_removed=[x for x in data if x >=lower and x <= upper]
print("outliers_removed is:{}" .format(len(outliers_removed)))
```

# interquartile range method

```python
# Identify outliers with interquartile range
from numpy.random import seed
from numpy.random import randn
from numpy import percentile
seed(1)

# generate univariate dataset
data= 5* randn(10000)+50

#calculate interquartile range
q25, q75=percentile(data, 25), percentile(data, 75)
iqr =q75-q25
print(q25, q75, iqr   

#calculate outlier cutoff
cut_off=iqr*1.5
lower,upper= q25-cut_off, q75+cut_off

# identify outliers
outliers= [x for x in data if x<lower or x> upper]
print("this is outliers:{}" .format (len(outliers)))

#remove outliers
outliers_removed= [ x for x in data if x>=lower and x<=upper]
print("this is removed outliers:{}" .format (len(outliers_removed)))
```
