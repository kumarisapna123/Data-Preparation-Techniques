# Feature Selection

```python
# in this section of data preparation we will learn about feature selection
```

```python
# according to me in a data set there are n no. of input features and 
# we have to reduce the input features to m where m is less than n 
# then the process of reducing the input features is known as features selection.
```

```python
# why do we need feature selection? 
# so by reducing the input features it enhances the performances of the machine learning model 
# it reduces the computational cost of the modeling.
```

feature selection methods based on output( targets)

```python
# it can be categorized as :
# supervised feature selection method:- In this method input variable is depend upon output variable ( like removing irrelevant variables)
# unsupervised feature selection method:- in this method input variable does not uses output variable ( basic data cleaning) 
```

supervised selection methods may further divided into three groups.

```python
# 1) intrinsic:- algorithms that performs automatic feature selection during training.
# 2) filter:- it selects subsets of input features based on the realtionship with the output variables.
# 3) wrapper:-  it searches for the subsets of input features that performs according to the machine learning model.
```
